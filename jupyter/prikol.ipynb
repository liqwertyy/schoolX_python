```python
import nltk
from nltk.tokenize import word_tokenize
```

```python

with open('book.txt', 'r') as file:
    text = file.read()
print(text)

```

```python
list(text)
```

```python
whitespace = nltk.tokenize.WhitespaceTokenizer()
print(whitespace.tokenize(text))
```

```python
wordpunct = nltk.tokenize.WordPunctTokenizer()
print(wordpunct.tokenize(text))
```

```python
PunktSentenceTokenizer = nltk.tokenize.PunktSentenceTokenizer()
print(PunktSentenceTokenizer.tokenize(text))
```

```python
Toktok = nltk.tokenize.ToktokTokenizer()
print(Toktok.tokenize(text))
```

```python
TreebankWordTokenizer = nltk.tokenize.TreebankWordTokenizer()
print(TreebankWordTokenizer.tokenize(text))
```

```python
import spacy
model = spacy.load('en_core_web_sm')
doc = model(text)
tokens = [token.text for token in doc]
tokens
```
