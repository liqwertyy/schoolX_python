```python
import spacy
```

```python
with open('1911posle-bala.txt', 'r', encoding='cp1251') as file:
    text6 = file.read()
```

```python
nlp = spacy.load('ru_core_news_sm')
doc = nlp(text6)
```

```python
sentences = list(doc.sents)[:100]
```

```python
for sentence in sentences:
    for token in sentence:
        print(token.text, token.pos_, token.dep_)
```

```python
counter = 0

for sentence in sentences:
    for token in sentence:
        if token.dep_ == 'ROOT':
            print(token.text)
            counter += 1
            if counter >= 100:
                break
    if counter >= 100:
        break
```
