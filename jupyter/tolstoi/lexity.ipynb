```python
with open('1852detstvo.txt', 'r', encoding='cp1251') as file:
    text1 = file.read()
with open('1868VoinaImir.txt', 'r' ,encoding='cp1251') as file:
    text2 = file.read()
with open('1878anna-karenina.txt', 'r', encoding= 'utf-8') as file:
    text3 = file.read()
with open('1885Ludi.txt', 'r', encoding='cp1251',) as file:
    text4 = file.read()
with open('1889Kreicerova.txt', 'r', encoding= 'utf-8') as file:
    text5 = file.read()
with open('1911posle-bala.txt', 'r', encoding='cp1251') as file:
    text6 = file.read()
```

```python
nlp = spacy.load('ru_core_news_sm')

doc = nlp(text1)


tokens = [token.text for token in doc if not token.is_punct and not token.is_stop]
words_filtered = nlp(" ".join(tokens))

word_freq = Counter(token.text for token in words_filtered if token.is_alpha)

num_adjectives = sum(1 for token in words_filtered if token.pos_ == "ADJ")

most_common_words = word_freq.most_common(20)

word_cloud = WordCloud().generate_from_frequencies(word_freq)
plt.imshow(word_cloud, interpolation='bilinear')
plt.axis("off")
plt.show()

print(f"Самые часто употребляемые слова: {most_common_words}")
print(f"Количество прилагательных: {num_adjectives}")
```
