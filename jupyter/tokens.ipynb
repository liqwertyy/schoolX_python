```python
#лексическе разнообразие

# import nltk
# from nltk.tokenize import ToktokTokenizer
# from collections import Counter

# tokenizer = ToktokTokenizer()

with open('Orwell1984.txt', 'r', encoding='utf-8') as file:
    textfull = file.read()

# tokens = tokenizer.tokenize(text)

# word_counts = Counter(tokens)

# for word, count in word_counts.items():
#     print(word, count)


#Лексичечкое разнообразие колобка(eng) 
# tokens = text.split() 
# word_count = Counter(tokens) 
# for word, count in word_count.items(): 
#     print(count, word) 




import string
def clean(textfull): 
    table = str.maketrans("", "", string.punctuation) 
    onlyWord = textfull.translate(table) 
    return onlyWord 
 
def diversity(textfull): 
    words = textfull.split() 
    uniqueWords = set(words) 

    print('всего слов:', len(words))
    print('количество уникальных слов:', len(uniqueWords)) 
    return len(uniqueWords) / len(words) 
 
onlyWord = clean(textfull) 
diversity = diversity(onlyWord) 
 
print('лексическое разнообразие:', diversity)

```

```python
class MyTokenizer:
    def __init__(self):
        pass

    def tokenize(self, text):
        words = text.split() 
        formatted_words = [word.capitalize() for word in words]  
        return formatted_words

tokenizer = MyTokenizer()

Mytext = "vjq ntrcn jq hfcrkflrf yt gjvtyzkfcm ye b kflyj PYTHON BAMBAMBAMBAMBA BIM BIMBIM"

tokens = tokenizer.tokenize(Mytext)

for token in tokens:
    print(token)
```
